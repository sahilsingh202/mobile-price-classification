##### 2. What is/are the insight(s) found from the chart?

Answer - The count plot shows that all four price categories contain almost the same number of mobile phones. This indicates that the dataset is balanced, with no single class dominating the others. A balanced target variable helps reduce bias during model training and improves the reliability of classification algorithms.

##### 3. Will the gained insights help creating a positive business impact?

Answer - A balanced dataset allows machine learning models to learn each price category equally, resulting in more accurate and reliable predictions. #### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation Matrix
corr = df.corr()

# Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap="coolwarm")

# Add Title and Axis Labels
plt.title("Correlation Heatmap", fontsize=16)
plt.xlabel("Features", fontsize=12)
plt.ylabel("Features", fontsize=12)

plt.show()
````

#### Chart - 15 - Pair Plot

````python
# Pair Plot visualization code

pair_data = df[['battery_power',
                'ram',
                'int_memory',
                'px_width',
                'price_range']]

sns.pairplot(
    pair_data,
    hue='price_range',
    diag_kind='hist',
    corner=True,
    palette='viridis'
)
````
